```python
# import libraries for exploration
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# option to show as many columns as possible
pd.set_option("display.max_rows", 101)
```

```python
# import datafiles
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

# create a list to loop over both
data = [train,test]
```

```python
# what does the dataset look like?
train.tail(20)
```

```python
# Data Cleaning: Correcting?

# Are there any "numeric" variables that seem odd in their distribution?
train.describe()
```

```python
# Data cleaning: completing

# what can we say about the amount of missing values in the data?
print("train:\n {}".format(train.isnull().sum()))
print("test:\n {}".format(test.isnull().sum()))

# most of the missing values are in cabin and in age as well as 2 in embarked.

print("there are {} samples in the training dataset and {} in the test dataset".format(train.shape[0], test.shape[0]))
```

```python
# Data Cleaning II: Completing

# fill age with median
# Fare with median Fare
# and Embarked with the mode

# (can be modified later)

for df in data:
    #df.fillna(df['Age'].median(), inplace=True) # save this for later
    df.fillna(df[['Fare']].median(), inplace=True)
    df.fillna(df['Embarked'].mode()[0],inplace=True)
    
# are there missing values left?
# what can we say about the amount of missing values in the data?
print("train:\n {}".format(train.isnull().sum()))
print("test:\n {}".format(test.isnull().sum()))
```

```python
# Data Cleaning: Creating

# can we create new variables from the data present?

# family size = number of siblings and spouses and children or parents
for df in data:
    df['FamilySize'] = df.SibSp + df.Parch + 1
    df['Title'] = df['Name'].apply(lambda name: name.split(', ')[1].split('.')[0].strip())

    
# stolen solution
# normalize the titles
standardized_titles = {
    "Capt":       "Officer",
    "Col":        "Officer",
    "Major":      "Officer",
    "Jonkheer":   "Royalty",
    "Don":        "Royalty",
    "Sir" :       "Royalty",
    "Dr":         "Officer",
    "Rev":        "Officer",
    "the Countess":"Royalty",
    "Dona":       "Royalty",
    "Mme":        "Mrs",
    "Mlle":       "Miss",
    "Ms":         "Mrs",
    "Mr" :        "Mr",
    "Mrs" :       "Mrs",
    "Miss" :      "Miss",
    "Master" :    "Master",
    "Lady" :      "Royalty"
}

for df in data:
    df['Title'] = df['Title'].map(standardized_titles)
# map the normalized titles to the current titles 
```

```python
for df in data:
    df['Age'] = df.Age.apply(lambda age: pd.to_numeric(age, errors = 'coerce'))
    
print(train.info())

# group by Sex, Pclass, and Title 
grouped = train.groupby(['Sex','Pclass', 'Title'])  
# view the median Age by the grouped features 
grouped.Age.median()

for df in data:
    df['Age'] = grouped.Age.apply(lambda age: age.fillna(age.median()))
# apply the grouped median value on the Age NaN

print(train.Age.isnull().sum())
```

```python
# Data cleaning: Converting

# convert all factor variables to dummies

train['Sex'] = train.Sex.map({"male": 0, "female":1})
# create the dummies
pclass_dummies = pd.get_dummies(train.Pclass, prefix = "Pclass")
title_dummies = pd.get_dummies(train.Title, prefix = "Title")
embarked_dummies = pd.get_dummies(train.Embarked, prefix = "Embarked")
# concatenate the dummies
train = pd.concat([train, pclass_dummies, title_dummies, embarked_dummies], axis = 1)

test['Sex'] = test.Sex.map({"male": 0, "female":1})
# create the dummies
pclass_dummies = pd.get_dummies(test.Pclass, prefix = "Pclass")
title_dummies = pd.get_dummies(test.Title, prefix = "Title")
embarked_dummies = pd.get_dummies(test.Embarked, prefix = "Embarked")
# concatenate the dummies
test = pd.concat([test, pclass_dummies, title_dummies, embarked_dummies], axis = 1)

```

```python
# did we get rid of all the nulls?
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
# what is the dispersion of age?
bins = np.arange(10,100,10)
totals = [sum(train.Age <= bin) for bin in bins]
survived = [sum((train.Age <= bin) & (train.Survived==1)) for bin in bins]
survived_bars = [i / j * 100 for i,j in zip(survived, totals)]
total_bars = [(j-i)/j*100 for i,j in zip(survived, totals)]

r = bins
barWidth = 8
names = ('Male', "Female")
plt.bar(r, survived_bars, color ='#b5ffb9', width = barWidth)
plt.bar(r, total_bars, bottom = survived_bars, color = "#e3e3e3", width = barWidth)
plt.xticks(r, bins)
plt.xlabel("Survived by age")
plt.show()

# the plot shows younger childeren with somewhat of a higher chance of survival
```

```python
#correlation heatmap of dataset
def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(14, 12))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':8 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)

correlation_heatmap(train)

# look through it to see if there are significantly highly correlated variables so they don't all go into the ML model
```

```python
# MODEL generation

# define features to use for the model:
target = ['Survived']
model_columns = ['Sex','Embarked_C', 'Embarked_Q', 'Embarked_S',
           'Age','Fare','FamilySize','Pclass_1', 'Pclass_2',
          'Pclass_3', 'Title_Master', 'Title_Miss','Title_Mr','Title_Mrs',
          'Title_Officer', 'Title_Royalty'] #pretty name/values for charts

# first split the training data in a training and a validation set

from sklearn.model_selection import train_test_split

training_x, validate_x, training_y, validate_y = train_test_split(train[model_columns], train[target], 
                                                                  test_size=.25, random_state=42)

print("train: {}\nvalidate: {}".format(len(training_x), len(validate_x)))
# splitting succesfull
```

```python
# fit a logistic regression and a random forest model
from sklearn.linear_model import LogisticRegression

# initialize it first:
lr = LogisticRegression(solver='liblinear')

# fit the lr to the data
lr.fit(training_x, training_y.values.ravel())

# predict on the validation set
predicted_y = lr.predict(validate_x)
```

```python
from sklearn import metrics

confusion_matrix = metrics.confusion_matrix(validate_y, predicted_y)
confusion_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(validate_y, predicted_y).round(2))
print("Precision:",metrics.precision_score(validate_y, predicted_y).round(2))
print("Recall:",metrics.recall_score(validate_y, predicted_y).round(2))
print("F-score:",metrics.fbeta_score(validate_y, predicted_y, 1).round(2))
```

```python
from sklearn.ensemble import RandomForestClassifier

# instantiate the Random forest classifier
rf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)

# fit the rf to the data
rf.fit(training_x, training_y.values.ravel())

# predict on the validation set
predicted_y = rf.predict(validate_x)
```
